```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/AnetaPopelova/czechitas-python-pro-datovou-analyzu-2024-09/refs/heads/main/data/Stats_survey.csv')
data.head()
```

```python
data.rename(
    columns={
        "Timestamp": "Timestamp",
        "Your Sex?": "Sex",
        "Your Matric (grade 12) Average/ GPA (in %)": "Matric_GPA",
        "What year were you in last year (2023) ?": "Year",
        "What faculty does your degree fall under?": "Faculty",
        "Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)": "Yearly_GPA",
        "Your Accommodation Status Last Year (2023)": "Accommodation",
        "Monthly Allowance in 2023": "Monthly_Allowance",
        "Were you on scholarship/bursary in 2023?": "Scholarship",
        "Additional amount of studying (in hrs) per week": "Study_Hours",
        "How often do you go out partying/socialising during the week? ": "Socialising_Frequency",
        "On a night out, how many alcoholic drinks do you consume?": "Drinks_Per_Night",
        "How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)": "Missed_Classes",
        "How many modules have you failed thus far into your studies?": "Modules_Failed",
        "Are you currently in a romantic relationship?": "In_Relationship",
        "Do your parents approve alcohol consumption?": "Parents_Approve",
        "How strong is your relationship with your parent/s?": "Relationship_Parents",
    },
    inplace=True,
)
data.head()
```

```python
graf1 = sns.histplot(data['Yearly_GPA'],
                     color='green',
                     edgecolor='black',
                     kde=True,
                    )

sns.set(style = 'dark')
plt.title('Histogram Yearly GPA', fontsize=20, fontfamily='Times New Roman')
plt.xlabel('Yearly GPA', fontsize = 12, fontfamily = 'Times New Roman')
plt.ylabel('Number of Students', fontsize = 12, fontfamily = 'Times New Roman')
```

```python
graf2 = sns.boxplot(x=data['Yearly_GPA'],
                    color = 'yellow',
                    )

graf2.set(title= 'Boxplot Yearly GPA',
          xlabel='Yearly GPA',
          )
```

```python
graf3 = sns.countplot(x = data['Sex'])
graf3.set(
    xlabel = 'Sex',
    ylabel = 'Count',
    title = 'Counts of Women and Men',
   )
```

```python
data.dropna(subset = ['Monthly_Allowance', 'Year'], inplace = True)

allowance_mapp = {
    "R 4001- R 5000": 4500,
    "R 5001 - R 6000": 5500,
    "R 6001 - R 7000": 6500,
    "R 7001 - R 8000": 7500,
    "R 8000+": 8000,
}

data["Monthly_Allowance"] = (
    data["Monthly_Allowance"].replace(allowance_mapp).astype(int)
)
```

```python
data[~data['Year'].isin(['Postgraduate'])]
```

Hypotéza 1.: Souvislost mezi frekvencí socializování a zmeškanými hodinami ve škole

```python
data.dropna(subset = ['Socialising_Frequency', 'Missed_Classes'], inplace = True)
missed_map = {
    '0':0,
    '1':1,
    '2':2,
    '3':3,
    '4+':4,
}
data["Missed_Classes"] = (
    data["Missed_Classes"].replace(missed_map).astype(int))

data_filtr = data[['Missed_Classes', 'Socialising_Frequency']]

ax = sns.barplot(
    data = data_filtr,
    x = 'Socialising_Frequency',
    y = 'Missed_Classes',
    hue = 'Socialising_Frequency',
    palette = 'dark:red',
    errorbar = None,
)

ax.set(
    xlabel = 'Socialising Frequency',
    ylabel = 'Missed Classes',
    title = 'Correlation between of socialising frequency and number of missed classes'
)
```

Hypotéza 2: Souvislost mezi ročním GPA a počtem zameškaných hodin

```python
miss_map = {
    '0':0,
    '1':1,
    '2':2,
    '3':3,
    '4+':4,
}

data["Missed_Classes"] = (
    data["Missed_Classes"].replace(miss_map).astype(int)
)

ax = sns.scatterplot(data = data,
    x = 'Missed_Classes',
    y = 'Yearly_GPA',
    hue = 'Sex',
    palette = 'bright',
)
ax.set(
    xlabel = 'Missed Classes',
    ylabel = 'Yearly GPA',
    title = 'Correlation between missed classes and yearly GPA'
)
```

Hypotéza 3: Souvislost mezi zameškanými hodinami ve škole a nárokem na stipendium

```python
data.dropna(subset = ['Scholarship', 'Missed_Classes'], inplace = True)

missed_map = {
    '0':0,
    '1':1,
    '2':2,
    '3':3,
    '4+':4,
}
data["Missed_Classes"] = (
    data["Missed_Classes"].replace(missed_map).astype(int))

ax = sns.barplot(
    data = data,
    x = 'Scholarship',
    y = 'Missed_Classes',
    errorbar = None,
    hue = 'Scholarship',
    palette = 'pastel'
)

for container in ax.containers:
    ax.bar_label(container, label_type="edge")

ax.set(
    ylabel = 'Missed Classes',
    title = 'Collaration between number of missed classes and scholarship',
)
```
